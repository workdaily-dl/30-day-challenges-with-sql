````python
%%sql variavel <<
SELECT A.TrackId, 
       C.Name AS artista, 
       A.Name AS musica, 
       COALESCE(A.Composer, "") AS compositor, 
       D.Name AS genero, 
       (A.Bytes / 1024) AS size_mb
FROM Track A, Album B, Artist C, Genre D 
WHERE A.AlbumId = B.AlbumId
AND   A.GenreId = D.GenreId
AND   B.ArtistId = C.ArtistId;
````
